```python
# --- setup ---
import numpy as np
import matplotlib.pyplot as plt
from colossus.lss import mass_function as mf
from colossus.cosmology import cosmology

cosmo = cosmology.setCosmology('planck18')  # sets global cosmology

# --- 1) Mass function dn/dlnM for M200c ---
def hmf_m200c(M, z, model='tinker08', q='dndlnM'):
    """
    M : array (Msun/h)
    z : scalar or array
    q : 'dndlnM' (per lnM) or 'dndM'
    returns number density [(Mpc/h)^-3 per lnM] or [(Mpc/h)^-3 per Msun/h]
    """
    return mf.massFunction(M, z, mdef='200c', model=model, q_out=q)

# example plot at two redshifts
M = 10**np.linspace(12.5, 15.8, 200)  # Msun/h
for z in (0.25, 1.0):
    phi = hmf_m200c(M, z, q='dndlnM')
    plt.loglog(M, phi, label=f'z={z}')
plt.xlabel(r'M$_{200c}$ [M$_\odot$/h]')
plt.ylabel(r'dn/dlnM [(Mpc/h)$^{-3}$]')
plt.legend(); plt.grid(alpha=0.3); plt.tight_layout()
plt.show()

# --- 2) Expected counts in mass bins over a sky area and redshift slice ---
def expected_counts_M200c(M_edges, z1, z2, A_deg2, model='tinker08', nz=60):
    """
    M_edges : mass bin edges (Msun/h), length B+1
    [z1,z2] : redshift slice
    A_deg2  : sky area (deg^2)
    returns M_mid (geom. mean), expected counts per bin
    """
    fsky = A_deg2 / 41252.96125
    lnM  = np.log(M_edges)
    Mmid = np.sqrt(M_edges[:-1] * M_edges[1:])
    dlnM = np.diff(lnM)

    # redshift shells
    z_edges = np.linspace(z1, z2, nz+1)
    z_mid   = 0.5*(z_edges[:-1] + z_edges[1:])

    # comoving volume of each shell (full-sky), then scale by f_sky
    Dc = cosmo.comovingDistance(z_edges)                 # Mpc/h
    dV_shell = (4.0*np.pi/3.0) * (Dc[1:]**3 - Dc[:-1]**3)  # (Mpc/h)^3
    dV = fsky * dV_shell                                  # (Mpc/h)^3 over your area

    # stack dn/dlnM over z and integrate
    dndlnM = np.array([hmf_m200c(Mmid, z, model=model, q='dndlnM') for z in z_mid])  # (nz,B)
    Nexp = (dndlnM * dV[:, None]).sum(axis=0) * dlnM  # expected counts per mass bin

    return Mmid, Nexp

# example: counts in 5000 deg^2 between z=0.2 and 0.4
M_edges = 10**np.linspace(14.0, 15.8, 10)  # Msun/h
Mmid, Nexp = expected_counts_M200c(M_edges, 0.2, 0.4, A_deg2=5000)

plt.step(M_edges[1:], Nexp, where='post')
plt.xscale('log'); plt.yscale('log')
plt.xlabel(r'M$_{200c}$ [M$_\odot$/h]'); plt.ylabel('Expected counts')
plt.grid(alpha=0.3); plt.tight_layout(); plt.show()

# # If you want Poisson realizations around the expectation:
# from scipy.stats import poisson
# N_sample = poisson.rvs(Nexp, random_state=123)
```

```python
import numpy as np
from colossus.lss import mass_function as mf
from colossus.cosmology import cosmology
from colossus.halo import concentration, mass_defs
from scipy.stats import lognorm
from colossus.lss import bias 

cosmo = cosmology.setCosmology("planck18")  # or your custom cosmology


# --- M200c -> M500c -> lambda(M500c, z) ---
def M200c_to_M500c(M200c, z, cmodel="diemer15"):
    c200 = concentration.concentration(M200c, "200c", z, model=cmodel)
    M500c, _, _ = mass_defs.changeMassDefinition(
        M200c, c200, z, "200c", "500c", profile="nfw"
    )
    return M500c


def lambda_of_M200c(M200c, z, lnA=3.74, B=1.22, C=0.0, D=0.0):
    """
    Simple mean scaling: ln  = lnA + B ln(M500c/1e14) + C ln(1+z) + D z
    (Set C,D as desired; here default is no explicit z-term.)
    """
    # M500c = M200c_to_M500c(M200c, z)
    # lnM = np.log(M500c/1e14)
    lnM = np.log(M200c / 3e14)  # Using M200c directly for simplicity
    ln_lambda = lnA + B * lnM + C * np.log(1.0 + z) + D * z
    return np.exp(ln_lambda)


def expected_counts_in_lambda_bins(
    lambda_edges,
    z1,
    z2,
    A_deg2,
    model="tinker08",
    bias_model="tinker10",
    nz=1,
    nM=600,
    Mmin=1e12,
    Mmax=1e16,
    ln_lambda_scatter=None,
    cmodel="diemer15",
):
    """
    Expected counts, mean mass, and mean bias per richness bin over [z1,z2] and area A_deg2.

    Parameters
    ----------
    lambda_edges : array, shape (B+1,)
        Bin edges in richness .
    z1, z2      : floats
        Redshift range.
    A_deg2      : float
        Sky area in deg^2.
    model       : str
        Mass function model for Colossus, e.g., 'tinker08'.
    bias_model  : str
        Halo bias model for Colossus, e.g., 'tinker10'.
    nz          : int
        Number of redshift shells.
    nM          : int
        Number of mass grid points (log-spaced; Msun/h).
    Mmin, Mmax  : floats
        Mass range in Msun/h (M200c).
    ln_lambda_scatter : float or None
        If given, _ln of lognormal P(|M,z). If None, no scatter.
    cmodel      : str
        Concentration model for mass-definition conversion (if you switch to M500c).

    Returns
    -------
    N_lambda : array, shape (B,)
        Expected counts per  bin.
    mean_mass : array, shape (B,)
        Mean M200c (Msun/h) of objects contributing to each  bin (NaN if empty).
    mean_bias : array, shape (B,)
        Mean halo bias of objects in each  bin (NaN if empty).
    """
    # Mass grid (Msun/h, 200c)
    M = np.logspace(np.log10(Mmin), np.log10(Mmax), nM)
    lnM = np.log(M)
    dlnM = np.gradient(lnM)

    # Redshift shells and comoving shell volumes over the survey area
    z_edges = np.linspace(z1, z2, nz + 1)
    z_mid = 0.5 * (z_edges[:-1] + z_edges[1:])
    fsky = A_deg2 / 41252.96125

    Dc = cosmo.comovingDistance(np.zeros_like(z_edges), z_edges, transverse=False)  # Mpc/h
    dV_shell = (4.0 * np.pi / 3.0) * (Dc[1:] ** 3 - Dc[:-1] ** 3)   # full-sky, (Mpc/h)^3
    dV = fsky * dV_shell                                            # survey volume per shell

    B = len(lambda_edges) - 1
    N_lambda = np.zeros(B, dtype=float)
    mass_num = np.zeros(B, dtype=float)   # for <M> numerator
    bias_num = np.zeros(B, dtype=float)   # for <b> numerator

    for i, z in enumerate(z_mid):
        # Mass function: d n / d ln M  [(Mpc/h)^-3 per lnM]
        phi = mf.massFunction(M, z, mdef="200c", model=model, q_out="dndlnM")

        # Counts in each mass bin for this shell
        dN_M = phi * dlnM * dV[i]   # shape (nM,)

        # Bias per mass in this shell
        bM = bias.haloBias(M, z, mdef="200c", model=bias_model)

        # Map mass bins to 
        lam = lambda_of_M200c(M, z)  # shape (nM,)

        if ln_lambda_scatter is None:
            # No scatter: each mass bin goes wholly to the  bin of its mean 
            N_lambda += np.histogram(lam, bins=lambda_edges, weights=dN_M)[0]
            mass_num += np.histogram(lam, bins=lambda_edges, weights=dN_M * M)[0]
            bias_num += np.histogram(lam, bins=lambda_edges, weights=dN_M * bM)[0]
        else:
            # With lognormal scatter: distribute each mass bin across  bins by P(|M)
            sigma = float(ln_lambda_scatter)
            mu = np.log(lam)  # mean ln  at this M,z

            # Vectorized CDF differences for all masses and bin edges
            # cdf has shape (nM, B+1); probs -> (nM, B)
            cdf = lognorm.cdf(lambda_edges[None, :], s=sigma, scale=np.exp(mu)[:, None])
            probs = cdf[:, 1:] - cdf[:, :-1]

            # Contributions to each -bin from each M-bin
            contrib = dN_M[:, None] * probs                    # (nM, B)
            N_lambda += np.sum(contrib, axis=0)
            mass_num += np.sum(contrib * M[:, None], axis=0)
            bias_num += np.sum(contrib * bM[:, None], axis=0)

    # Means (guard against empty bins)
    with np.errstate(invalid="ignore", divide="ignore"):
        mean_mass = mass_num / N_lambda
        mean_bias = bias_num / N_lambda

    # Put NaN where bins are empty
    empty = N_lambda <= 0
    mean_mass[empty] = np.nan
    mean_bias[empty] = np.nan

    return N_lambda, mean_mass, mean_bias
```

```python
# Define richness bins
lambda_edges = np.array([10, 20, 30, 45, 70, 120, 220, 300])

# Counts in 5000 deg^2 between z=0.2 and 0.4, no scatter
N_lam, mean_mass, mean_bias = expected_counts_in_lambda_bins(
    lambda_edges, 0.2, 0.65, A_deg2=5000
)

# Same but with lognormal scatter _ln = 0.25
N_lam_scatter, mean_mass_scatter, mean_bias_scatter = expected_counts_in_lambda_bins(
    lambda_edges, 0.2, 0.65, A_deg2=5000, ln_lambda_scatter=0.18
)

print("Counts per -bin (no scatter):", N_lam, mean_mass, mean_bias)
print("Counts per -bin (_ln=0.18):", N_lam_scatter, mean_mass_scatter, mean_bias_scatter)
```
